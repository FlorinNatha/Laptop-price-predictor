```python
# function ekk haduwa comany eke thina laptop count eka 10 t wada adu ewa other kiyan category ekta damma
def add_company(inpt):
    if inpt == 'Samsung' or inpt == 'Razer' or inpt == 'Mediacom' or inpt == 'Microsoft' or inpt == 'Xiaomi' or inpt == 'Vero' or inpt == 'Chuwi' or inpt == 'Google' or inpt == 'Fujitsu' or inpt == 'LG' or inpt == 'Huawei':
        return 'other'
    else:
        return inpt

data['Company'] = data['Company'].apply(add_company)
```

```python
#ScreenResolution column eke 'Touchscreen' thinwad kiyla balnwa. thinawa nm 1 kiyla danwa naththm 0
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
#ScreenResolution column eke 'IPS' thinwad kiyla balnwa. thinawa nm 1 kiyla danwa naththm 0
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
#aluth column ekk haduwa 'cup_name' kiyala ekedi 'Cpu' column eke apu values wal mulma kotas 3 gnnawa
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name == 'Intel Core i7' or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'other'

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
# gpu_name wla tina "ARM" data eka delete kara
data = data[data['gpu_name'] != 'ARM']
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return 'Linux'
    else:
        return 'Other'

data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data = data.drop(columns=['laptop_ID', 'Inches', 'Product', 'ScreenResolution', 'Cpu', 'Gpu'])
```

```python
data.head()
```

```python
# okkoma data set ekam one-hot encoding karnwa (numerical kara)
data = pd.get_dummies(data)
```

```python
data.head(2)
```

```python
data.shape
```

```python
#........................Model Building..........
```

```python
# x- Price_euros column eka nathuwa ithiri tika 
# y- Price_euros column eka
X = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
!pip install scikit-learn
```

```python
#train waltai test wlati data bedanwa
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
#train and test walat kochchra data bedilad balnawa
X_train.shape, X_test.shape
```

```python
# meka Regression proble ekk ; so algorithms : linear regression, laso regression, decision tree , Random forese
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
#LinearRegression Algorithm
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
#Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
#RandomForest Regressor
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
#hyper parameter tuning
from sklearn.model_selection import GridSearchCV

#'n_estimators' and 'criterion' parameters 2 use karanwa(RandomForestRegressor wla tina oni tharam parameters use karnna puluwn
parameters = {'n_estimators':[10, 50, 100],
               'criterion':['squared_error', 'absolute_error', 'poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

#model eka train krnwa
grid_fit = grid_obj.fit(X_train, y_train)

# sudusuma parameters tina model ekk hadagnwa
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test, y_test)
```

```python
# mee model eka use karala website ekk hadanwa. so mee model eka save karagnn oni
import pickle
with open('predictor.pickel', 'wb') as file:
    pickle.dump(best_model, file)
```
